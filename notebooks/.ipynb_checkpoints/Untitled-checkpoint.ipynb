```python
import tensorflow as tf
import numpy as np
import os
import sys
import datetime
import time
import yaml
import matplotlib.pyplot as plt

# to be able to show colored images
sys.path.insert(1, r'C:\Users\shams\PhD\segmentation\scripts\preprocess')
import reading_image as r_i
```

```python
train_data_path_image= r'C:\Users\shams\PhD\segmentation\dataset\large_gis\png_augmented_360\test_dest\train\image'
train_data_path_label= r'C:\Users\shams\PhD\segmentation\dataset\large_gis\png_augmented_360\test_dest\train\label'
batch_size = 8
```

```python
train_image_datagen = tf.keras.preprocessing.image.ImageDataGenerator(
    preprocessing_function=lambda x: x / (2 ** 8 - 1),
)
train_mask_datagen = tf.keras.preprocessing.image.ImageDataGenerator(
)

train_image_generator = train_image_datagen.flow_from_directory(train_data_path_image, class_mode=None,
                                                                target_size=(512, 512), seed=12,
                                                                batch_size=batch_size)

train_mask_generator = train_mask_datagen.flow_from_directory(train_data_path_label, class_mode=None,
                                                              target_size=(512, 512), seed=12,
                                                              batch_size=batch_size,
                                                              color_mode='grayscale')
train_generator_dictionary = zip(train_image_generator, train_mask_generator)
```

```python
number_images = 30
```

```python
for idx, (img_el, mask_el) in enumerate(train_generator_dictionary):
    if idx > number_images:
        break

    print(f'img shape {img_el.shape}')
    print(f'mask shape {mask_el.shape}')
    fig, ax = plt.subplots(2, 2)
    
    ax[0, 0].imshow(img_el[0])
    ax[0, 1].imshow(r_i.get_colored_label(np.squeeze(mask_el[0])))
    
    ax[1, 0].imshow(img_el[1])
    ax[1, 1].imshow(r_i.get_colored_label(np.squeeze(mask_el[1])))
    
    plt.show()
```
