```python
# changing 5 with 0 and 0 with 5
def swap_two_values(element, val1 = 5, val2= 0):
    if element == val1:
        return val2
    elif element == val2:
        return val1
    else:
        return element
    
def change_noise(element):
    if element < 0:
        return 0
    elif element > 5:
        return 5
    else:
        return element

# def array_swap(arr):
#     return np.array([swap_two_values(arr_el) for arr_el in arr])

# def label_scipy_noise_handle(arr):
#     return np.array([change_noise(arr_el) for arr_el in arr])

# vectorization of the functions using numpy
vectorised_noise_scipy_handle = np.vectorize(change_noise)
vectorised_swap = np.vectorize(swap_two_values)
```

```python
# write arry to raster using Arilds code
def write_array_to_raster( output_filepath, array, geo_transform, projection):
    """
    Writes the given array to a raster image
    :param output_filepath: The output file
    :return: Nothing
    """
    bands = array.shape[-1] if len(array.shape) > 2 else 1
    driver = gdal.GetDriverByName("GTiff")
    raster = driver.Create(output_filepath, array.shape[1], array.shape[0],
                  bands, gdal.GDT_Int16)
    raster.SetGeoTransform(geo_transform)
    raster.SetProjection(projection)
    # Write multiple bands
    if len(array.shape) > 2:
        for band in range(array.shape[-1]):
            raster.GetRasterBand(band + 1).WriteArray(array[:,:,band])
    # Write single band
    else:
        raster.GetRasterBand(1).WriteArray(array)
    raster = None
```

```python
# this function gives us a pair of label-image of corresponding folder
def get_img_lbl_pair(image_addr, label_addr, river_name, image_number):
    # get the river path
    image_addr_river = os.path.join(image_addr, river_name)
    label_addr_river = os.path.join(label_addr, river_name)
    
    # get all the files in the path in a list
    label_addr_river_list = os.listdir(label_addr_river)
    image_addr_river_list = os.listdir(image_addr_river)
    
    # select on label based on number and remover label at the beginning
    single_label_addr = label_addr_river_list[image_number]
    single_image_addr = single_label_addr.replace('label','')
    
    # look for the same name in the images, if not raise an error
    if not single_image_addr in image_addr_river_list:
        raise Exception(f'the label and image {single_image_addr} are not the same')
    
    return os.path.join(image_addr_river, single_image_addr), os.path.join(label_addr_river, single_label_addr)
```

```python
# get image, corresponding label, transformation and projection
def get_images(img_path, lbl_path):
    # Load image
    image_ds = gdal.Open(img_path)
    geo_transform = image_ds.GetGeoTransform()
    projection = image_ds.GetProjection()
    image_matrix = image_ds.GetRasterBand(1).ReadAsArray()
    image_ds = None

    # Load label
    label_ds = gdal.Open(lbl_path)
    if label_ds.GetGeoTransform() != geo_transform:
        raise Exception(f"The geo transforms of image {image_filepath} and label {label_filepath} did not match")
    label_matrix = label_ds.GetRasterBand(1).ReadAsArray()
    label_ds = None
    return image_matrix, label_matrix, geo_transform, projection

# image_matrix, label_matrix, geo_transform, projection = get_images(image_filepath, label_filepath)
```

```python
def rotate_scipy(angle, img, lbl):
    from scipy import ndimage, misc
    return ndimage.rotate(img, angle, reshape=False), ndimage.rotate(lbl, angle, reshape=False)

def rotate_skimage(sk_angel, img, lbl):
    from skimage.transform import rotate as sk_rotate
    return sk_rotate(img, sk_angel), sk_rotate(lbl, sk_angel, preserve_range=True).astype(np.uint8)
```

```python
def make_folder(folder_name):
    if not os.path.exists(folder_name):
        os.makedirs(folder_name)
```

```python
def get_augment_rotate_double(image_matrix, label_matrix_changed, small_image_size, selection_lim, selected_class,
                       rotation_angle_list, large_image_file_name, labeled_percentage=0.1, overlap_percentage=0.5, method='scipy'):
    # check if the picture has more than 100 pixel of selected calss (for example gravel)
    arr_tmp = np.zeros((1, 6))
    unique_tmp, count_tmp = np.unique(label_matrix_changed, return_counts=True)
    arr_tmp[0, unique_tmp] = count_tmp
    if arr_tmp[0, selected_class] < 1000:
        return False, False, False, False
    
    # csv list consists of [large image name, angle, xcenter, ycenter]
    csv_list = []
    image_list = []
    label_list = []
    global_counter = 0
    not_useful_count = 0
    for angle in rotation_angle_list:
        # set the angle
        if method == 'scipy':
            tmp_large_image, tmp_large_label = rotate_scipy(angle, image_matrix, label_matrix_changed)
        elif method == 'skimage':
            tmp_large_image, tmp_large_label = rotate_skimage(angle, image_matrix, label_matrix_changed)
        else:
            raise ('method is not defined')

        # make a mask of large image
        # mask_large controls the unknown class
        masked_large = tmp_large_label > 0
        # overlap_large controls the overlap of sampling
        overlap_large = tmp_large_label > -1
        
        x, y = np.where(tmp_large_label == selected_class)
        print(f'x shape is {x.shape[0]}')

        for select_no in range(x.shape[0] * 2):
            # select one pixel to be the center of the frame
            # in order to avoid forever loops, we set a counter
            while_counter_lim = x.shape[0]
            while_counter = 0
            while True:
                # check the lim of the while
                if while_counter > while_counter_lim:
                    break
                while_counter += 1

                pixel_idx = np.random.randint(0, x.shape[0])
                if (small_image_size / 2) < x[pixel_idx] < tmp_large_label.shape[0] - (
                        small_image_size / 2) \
                        and (small_image_size / 2) < y[pixel_idx] < tmp_large_label.shape[1] - (
                        small_image_size / 2):
                    break
            # center pixel
            ctr_x = x[pixel_idx]
            ctr_y = y[pixel_idx]
            # range of frame work
            x_min = int(ctr_x - (small_image_size / 2))
            x_max = int(ctr_x + (small_image_size / 2))
            y_min = int(ctr_y - (small_image_size / 2))
            y_max = int(ctr_y + (small_image_size / 2))
            # check if more than 50 percent of the labels are masked as True
            label_true_percent = (masked_large[x_min:x_max, y_min:y_max].sum()) / (
                masked_large[x_min:x_max, y_min:y_max].size)
            overlap_true_percent = (overlap_large[x_min:x_max, y_min:y_max].sum()) / (
                overlap_large[x_min:x_max, y_min:y_max].size)
            # get the image
            tmp_small_label = tmp_large_label[x_min:x_max, y_min:y_max]
            tmp_small_image = tmp_large_image[x_min:x_max, y_min:y_max]
            # if the frame is on the right spot
            if (label_true_percent > labeled_percentage) and (overlap_true_percent > overlap_percentage):
                # change 6 to 5 and -1 to 0
                tmp_small_label = vectorised_noise_scipy_handle(tmp_small_label)
                # if there exist a label that is less than 1% of total pixels, it should be changed to the nearest
                # neighboar TODO
                image_list.append(tmp_small_image)
                label_list.append(tmp_small_label)
                csv_list.append([global_counter, large_image_file_name, angle, ctr_x, ctr_y])
                global_counter += 1
                # change the overlap mask to indicate that this pixels are already taken
                overlap_large[x_min:x_max, y_min:y_max] = False
            # if frame is not on the right spot
            else:
                not_useful_count += 1
    print(f'there were {not_useful_count} not useful images here')
    if len(image_list) == 0:
        return False, False, False, False
    return image_list, label_list, csv_list, masked_large, overlap_large
```

```python
def save_augmented_list(img_list, folder_addr, large_image_name, geo_transform, projection):
        for idx, label_element in enumerate(img_list):
            label_name = f'{large_image_name}_{idx}.tif'
            label_addr = os.path.join(folder_addr, label_name)
            # save label
        #     imageio.imwrite(label_addr, label_element) this one can work very well to save numpy array but I used GDAL to save the images
            write_array_to_raster(output_filepath=label_addr, array=label_element ,geo_transform=geo_transform, projection=projection)
```

```python
def analyzing_class(label_list, class_name_dict):
    image_class_no_list = []
    for label_el in label_list:
        uniq, count = np.unique(label_el, return_counts=True)
        # check if they have all 0 to 5
        for label_element in range(6):
            if not label_element in uniq:
                uniq = np.append(uniq, label_element)
                count = np.append(count, 0)
        image_class_no_list.append(dict(zip(uniq, count)))
    image_class_df = pd.DataFrame(image_class_no_list)
    image_class_numpy = image_class_df.to_numpy()
    image_class_df = (pd.DataFrame(image_class_numpy / (image_class_numpy.sum(axis=1))[:,None] * 100).round(2))
    return image_class_df.rename(columns=class_name_dict)

# a better version of algorithm above
def get_class_analysis(lbl_list, class_dict):
    image_range = len(lbl_list)
    arr_tmp = np.zeros((image_range, 6))
    for i in range(image_range):
        unique_tmp, count_tmp = np.unique(lbl_list[i], return_counts=True)
        arr_tmp[i,unique_tmp] = count_tmp
    return pd.DataFrame((arr_tmp / (arr_tmp.sum(axis=1))[:,None] * 100).round(2)).rename(columns=class_dict)
```
