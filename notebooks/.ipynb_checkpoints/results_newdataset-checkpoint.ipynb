# new dataset results

show the csv and miou of different experiments

```python
import pandas as pd
import numpy as np
import os
```

```python
def show_river_results(base_path, river_name):
    miou_path = os.path.join(base_path, river_name, 'predictions', 'miou.csv')
    conf_path = os.path.join(base_path, river_name, 'predictions', 'confusion_matrix_test.csv')
    miou_df = pd.read_csv(miou_path)
    conf_df = pd.read_csv(conf_path, names=['water', 'gravel', 'vegetation', 'farmland', 'hum_cons'])
    conf_df = conf_df.div(conf_df.sum(axis=1), axis=0).round(2) * 100
    print(f'MIoU is {miou_df}')
    return conf_df

def show_river_results_with_unknown(base_path, river_name, prediction_name='predictions'):
    miou_path = os.path.join(base_path, river_name, prediction_name, 'miou.csv')
    conf_path = os.path.join(base_path, river_name, prediction_name, 'confusion_matrix_test.csv')
    miou_df = pd.read_csv(miou_path)
    conf_df = pd.read_csv(conf_path, names=['water', 'gravel', 'vegetation', 'farmland', 'hum_cons','unknown'])
    conf_df = conf_df.div(conf_df.sum(axis=1), axis=0).round(2) * 100
    print(f'MIoU is {miou_df}')
    return conf_df
```
