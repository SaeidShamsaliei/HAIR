### Sobel in python

```python
def show_sobel(image):
    fig = plt.figure(figsize = (10,10))
    plt.gray()  # show the filtered result in grayscale
    ax1 = fig.add_subplot(121)  # left side
    ax2 = fig.add_subplot(122)  # right side

    result = ndimage.sobel(image)

    ax1.imshow(image)
    ax2.imshow(result)
    plt.show()
```

```python
def sobel_filter(img, sigma=5):
    """
    performs sobel after removed the noise using gaussian filter
    """
    # remove noise
    img = ndimage.gaussian_filter(img, sigma=sigma)
    # apply laplacian
    result = ndimage.sobel(img)
    
    return result
```

### laplacian

```python
def show_laplacian(image):
    fig = plt.figure(figsize = (10,10))
    plt.gray()  # show the filtered result in grayscale
    ax1 = fig.add_subplot(121)  # left side
    ax2 = fig.add_subplot(122)  # right side

    result = ndimage.laplace(image)

    ax1.imshow(image)
    ax2.imshow(result)
    plt.show()
```

```python
def laplacian_filter(img, sigma=5):
    """
    performs laplacian after removed the noise using gaussian filter
    """
    # remove noise
    img = ndimage.gaussian_filter(img, sigma=sigma)
    # apply laplacian
    result = ndimage.laplace(img)
    
    return result
```

```python
image_n = 8
original_image = images[image_n].data

show_laplacian(original_image)
```

```python
image_n = 8
original_image = images[image_n].data

result1 = ndimage.laplace(original_image)
result2 = laplacian_filter(original_image)

fig = plt.figure(figsize = (10,10))
plt.gray()  # show the filtered result in grayscale
ax1 = fig.add_subplot(131)  # left side
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

ax1.imshow(result1)
ax2.imshow(result2)
ax3.imshow(original_image)
```
