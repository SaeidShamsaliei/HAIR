```python
raster_name = 'merged_output'
file_path = f'/home/saeid/phd/segmentation/dataset/TestPrediction/tmp_output/{raster_name}.tif'
```

```python
def raster_to_numpy(file_path_val):
    sourceRaster = gdal.Open(file_path_val)
    sr_proj = sourceRaster.GetProjection()
    raster_proj = osr.SpatialReference()
    raster_proj.ImportFromWkt(sr_proj)
    band = sourceRaster.GetRasterBand(1) 
    bandArray = band.ReadAsArray()
    sourceRaster = None
    return bandArray
```

```python
numpy_raster = raster_to_numpy(file_path)
```

```python
plt.imshow(numpy_raster)
plt.show()
```

```python
img = numpy_raster
kernel = np.ones((10,10),np.uint8)
erosion = cv2.erode(img,kernel,iterations = 1)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel, iterations = 3)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel, iterations = 3)
open_close = cv2.morphologyEx(opening, cv2.MORPH_CLOSE, kernel)
close_open = cv2.morphologyEx(closing, cv2.MORPH_OPEN, kernel)
```
